### Basic idea

The power method is a simple yet highly effective iterative algorithm for approximating the dominant eigenvalue and corresponding to the eigenvalue of largest modulus. Beginning with a random initial vector, the method repeatedly applies the matrix and normalizes the result, producing progressively better approximations. Despite being nearly a century old, the power method remains central in modern numerical algorithms. Its enduring relevance stems from two key strengths: exceptional computational efficiency and remarkable robustness. These properties make the power method a practical and reliable tool for computing the dominant eigenvalue in large-scale problems.

### Implementation

1. Define the power method with input variables 'matrix', the matrix A; 'x0', the initial vector; 'iter', the maximum number of iterations; and 'TOL', the tolerance of the change. If 'TOL' is larger than the change in between the predicted consequential eigenvalues the method stops.

2. We use a for loop for every iteration.

3. We calculate the next iteration of the power method and normalize the solution vector.

4. We calculate the dominant eigenvalue. 

5. Check if the eigenvalue is converging to the actual eigenvalue. 
 

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Define the power method

```python
def power_method(matrix, x0, iter, TOL = 1e-6):
    '''
    Finds the largest eigenvector using the power method and scaling
    Inputs:
        matrix - the matrix
        x0 - the initial vector
        iter - maximum number of iterations
        TOL - maximum error
    '''
    # store the largest eigenvalue approximations 
    rayleighs = []

    for i in range(iter):
        # power method iteration
        x0 = matrix @ x0

        #power method normalization 
        x0 = 1/max(x0)*x0

        # calculate the largest eigenvalue
        rayleigh = (x0 @ matrix @ x0)/(x0 @ x0)
        rayleighs.append(rayleigh)

        # if the eigenvalue already converged we can stop the algorithm 
        if np.abs(rayleigh - rayleighs[i-1]) < TOL and i > 3:
            break
    
    return x0, rayleighs
```

### 2x2 matrix implementation

#### Converging eigenvector

```python
matrix = np.array([[4, -5], [2, -3]])
x0 = np.ones(2)
iter = 50
x, rayleighs = power_method(matrix, x0, iter)
```

```python
iterations = [i for i in range(len(rayleighs))]
print(rayleighs)
plt.plot(iterations, rayleighs)
```

#### Non-converging eigenvector

```python
matrix = np.array([[-3, 0], [1, 3]])
x, rayleighs = power_method(matrix, x0, iter)
```

```python
iterations = [i for i in range(len(rayleighs))]
print(rayleighs)
plt.plot(iterations, rayleighs)
```

### 3x3 implementation

```python
matrix = np.array([[2, 3, 1], [0, -1, 2], [0, 0, 3]])
x0 = np.ones(3)
x, rayleighs = power_method(matrix, x0, iter)
```
